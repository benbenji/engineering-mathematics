```python
from sympy import *
```

```python
x = symbols('x')
y = x**2*exp(-1/2*x)
y1 = y.diff(x)
solve(y1>0,x)
```

```python
solve(y1<0,x)
```

```python
y = (2+7*x)/sqrt(3+5*x+3*x**2)
y1 = y.diff(x)
solution = solve(y1,x)
```

```python
solution[0]
```

```python
plot(y)
```

```python
y.subs(x,solution[0]).evalf()
```

```python
y2 = y1.diff(x)
y2.subs(x,solution[0]).evalf()
```
