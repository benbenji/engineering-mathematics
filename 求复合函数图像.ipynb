```python
def complexsin_f(n):
    z = sin(x)
    for i in range(n-1):
        z = sin(z)
    return z
```

```python
x = symbols('x')
z = complexsin_f(5)
z1 = complexsin_f(10)
z2 = complexsin_f(30)
z4 = complexsin_f(1)
```

```python
plot(z,z1,z2,z4,(x,-6,6))
z2
```

```python
x = symbols('x')
def complex_f1(n):
    global x
    for i in range(n):
        x = x/sqrt(1+x**2)
    return x
z1 = complex_f1(1)
z2 = complex_f1(3)
plot(z1,z2)
```
